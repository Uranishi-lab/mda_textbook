#### **リストのリスト（2次元リスト）**

**ソースコード6.16** 2次元リスト

````python
score = [[40, 60, 80], [60, 75, 85]]  # ２次元リスト
print(score[0])                     # 生徒Aの成績
print(score[1])                     # 生徒Bの成績
print(score[0][1])                  # 生徒Aの数学の成績
print(score[1][2])                  # 生徒Bの英語の成績
````

**ソースコード6.17** 2次元リストの活用

````python
score = [[40, 60, 80], [60, 75, 85]]    # ２次元リスト（i行j列）
for i in range(2):                      # range(2)は[0, 1]を指定しているのと同じ
    sum = 0                             # i人目の生徒の合計点を入れるsumを0にする
    for j in range(3):                  # range(3)は[0, 1, 2]を指定しているのと同じ
        sum = sum + score[i][j]         # 列を0,1,2と進めて加えていく
    print(sum, sum/3)                   # 各生徒毎に合計点と平均点を表示
````
